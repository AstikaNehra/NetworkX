```python
import networkx as nx
import pandas as pd
```

```python
G=nx.Graph() #empty graph
```

```python
G.add_node(1) 
nx.draw(G)
#create one node
```

```python
G.add_nodes_from([2,3,4,5])
nx.draw(G,with_labels=True)
#adding labels of each node
```

```python
H=nx.path_graph(10)
#add 10 nodes to a graph
nx.draw(H,with_labels=True)
```

```python
G.add_edge(1,2)
e=(2,3)
G.add_edge(*e)
nx.draw(G,with_labels=True)
```

```python
G.add_edges_from([(1,3),(1,4)])
nx.draw(G,with_labels=True)
```

```python
df=pd.read_csv("1999-1-wrsn.tsv",sep="\t",header=None)
df.head(n=5)
```

```python
G=nx.from_pandas_edgelist(df,0,1,edge_attr=2,create_using=nx.MultiDiGraph())
H=nx.from_pandas_edgelist(df,0,1,edge_attr=2,create_using=nx.Graph())
```

```python
#circular layout
nx.draw(G, pos=nx.circular_layout(G))
```

```python
#spring layout
nx.draw(G,pos=nx.spring_layout(G),node_color="black",node_size=5,arrow_size=5)
```

```python
#random layout
nx.draw(G,pos=nx.random_layout(G),node_color="black",node_size=5,arrow_size=5)
```

```python
#connected path
subgraphs=list(nx.connected_component_subgraphs(H,copy=False))
```

```python
for subgraph in subgraphs:
    if len(subgraphs)==1:
        print("only 1 connected component")
        nx.draw(subgraph)
    else:
        print("found subgraph")
        nx.draw(subgraph)
```

```python
nx.shortest_path(G,777,40989,weight=2)
#find the shortest path between 777 and 40989
```
